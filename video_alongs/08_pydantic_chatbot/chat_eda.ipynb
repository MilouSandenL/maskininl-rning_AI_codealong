# Explore chatting with pydanticAI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv

# GEMINI_API_KEY
load_dotenv()

chat_agent = Agent(
    "google-gla:gemini-2.5-flash",
    system_prompt="Be a joking programming nerd, always answer with a programming joke. Also add in some emojis to make it funnier.",
)

chat_agent
```

```python
result = await chat_agent.run("hello")
result
```

```python
result.output
```

```python
# have no memory by default

result = await chat_agent.run("What did i ask you first?")
result.output
```

```python
result.all_messages()
```

### add memory to the agent

```python
result = await chat_agent.run("What is the weather?")
result
```

```python
result2 = await chat_agent.run("What did i ask you first?", message_history=result.all_messages())
result2.output
```
